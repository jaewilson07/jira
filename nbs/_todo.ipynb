```python
from requests import request


def looper(loop_fn, auth: JiraAuth, offset = 0, limit = 50, debug_loop: bool = False, **kwargs ):
    final_array = []
    keep_looping = True

    while keep_looping:

        if debug_loop:
            print({
                'offset' : offset,
                'limit' : limit,
                'fn_name' : loop_fn.__name__
            })

        new_array = loop_fn(auth = auth, offset = offset, limit = limit,  **kwargs)

        if not new_array:
            break
    
        final_array += new_array
        offset += limit

    return final_array

def get_boards(auth: JiraAuth,
    offset= 0, limit = 50, 
    ):

    url = f"https://{auth.instance}/rest/agile/1.0/board"

    params = {'offset': offset, 'limit': limit}

    headers = auth.generate_headers()

    res = request(method = 'get', url = url, headers = headers, params = params)

    data = res.json()

    return data['values']

# looper(loop_fn = get_boards , auth = jira_auth, offset = 0 , limit = 50, debug_loop= True)

# boards = get_boards(headers)

# [board for board in boards]
```
